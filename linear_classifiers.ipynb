## Proofs of SVM

### How do we find the hyper plane?

Our hyperplane in vector form:

$$
	w^T x - b = 0
$$

$w^T$: a normal vector to our plane
$x$: our point space
$b$: how far the plane is from the origin

### TL:DR; Derivation

$$
	y_i * (\bold{w \cdot x} + b) \le 1
$$

mathematically

$$
	\bold{w} = \sum a_i y_i \bold{x_i}
$$

$$
	\sum a_i - \frac{1}{2} \sum_i \sum_j a_i a_j y_i y_j \bold{x_i \cdot x_j}
$$

we are wanting to minimise the final function.



```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python
X = np.array([1, 5, 1.5, 8, 1, 9, 7, 8.7, 2.3, 5.5, 7.7, 6.1])
y = np.array([2, 8, 1.8, 8, 0.6, 11, 10, 9.4, 4, 3, 8.8, 7.5])
```

```python
plt.scatter(X, y)
plt.show()
```

```python
training_X = np.vstack((X, y)).T
training_y = [0, 1, 0, 1, 0, 1, 1, 1, 0, 0, 1, 1]
```

```python
clf = svm.SVC(kernel='linear', C=1.0)
```

```python
clf.fit(training_X, training_y)
```

```python
# get the weight values for the linear equation from the trained SVM model
w = clf.coef_[0]

# get the y-offset for the linear equation
a = -w[0] / w[1]

# make the x-axis space for the data points
XX = np.linspace(0, 13)

# get the y-values to plot the decision boundary
yy = a * XX - clf.intercept_[0] / w[1]

# plot the decision boundary
plt.plot(XX, yy, 'k-')

# show the plot visually
plt.scatter(training_X[:, 0], training_X[:, 1], c=training_y)
plt.show()
```
